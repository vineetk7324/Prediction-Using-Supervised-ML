AIM : TO PREDICT THE SCORE OF A STUDENT STUDIES FOR 9.25HOURS/DAY


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
print("STUDENT SCORES DATASET")
info = pd.read_csv('Student_scores.csv')
info.head()
```

```python
print('Line Graph')
info.plot(x='Hours',y='Scores')
plt.xlabel('HOURS').set_color('white')
plt.ylabel('SCORES').set_color('white')
plt.show()

print('Dotted Graph')
info.plot(x='Hours',y='Scores',style='o',color='red')
plt.xlabel('HOURS').set_color('white')
plt.ylabel('SCORES').set_color('white')
plt.show()
```

*From the above graph it is clearly visible that there is a pattrn of positive linearity.*

```python
x=info['Hours']   #feature column
y=info['Scores']  #target column
```

```python
import warnings
warnings.filterwarnings('ignore')

import seaborn
plt.xlabel('HOURS').set_color('white')
plt.ylabel('SCORES').set_color('white')
seaborn.regplot(x,y,color='red')
```

----------------------------*Forming - Training & Testing Data*-----------------------------

```python
a=info.describe()
print(a)
print("--------------------------------")
b=info.shape
print(b)
print("--------------------------------")
c=info.columns
print(c)
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(info,test_size=0.25,random_state=123)
```

```python
train_x=train.drop('Scores',axis=1)
train_y=train['Scores']
```

```python
test_x=train.drop('Scores',axis=1)
test_y=train['Scores']
```

---------------------------*Training Algorithm*----------------------------

```python
from sklearn.linear_model import LinearRegression 
linear=LinearRegression()
```

```python
linear.fit(train_x, train_y)
```

```python
predict=linear.predict(test_x)
```

-------------------------Actual Value vs Predicted Value------------------------------

```python
a = pd.DataFrame({'Actual':test_y,'Predicted':predict})
a
```

--------------What will be the predicted score if a student studies for 9.25hours/day ?------------

```python
hours=[9.25]
our_predict=linear.predict([hours])
print("HOURS STUDIED = {}".format([hours]))
print("PREDICTED SCORE = {}".format(our_predict[0]))
```

-----------------Evaluating The Algorithm-----------------

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
r2_score(test_y,predict)
```

```python
mean_squared_error(test_y,predict,squared=False)
```

```python
mean_absolute_error(test_y,predict)
```
